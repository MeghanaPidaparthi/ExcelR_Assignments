```python
region_sales = data.groupby('Region')['Sales'].sum()
region_sales #alows you to split data into groups,apply functions to each group independently and then combine the results.
```

```python
#finding the most sold product
data.groupby("Product")["Quantity"].sum()
```

```python
# Compute the average profit margin for each product
data["profit margin (%)"] = (data["Profit"]/data['Sales'])*100
avg_profit_margin= data.groupby("Region")['profit margin (%)'].mean()
avg_profit_margin
```
