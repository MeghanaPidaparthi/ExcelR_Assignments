```python
data = pd.read_csv('Day_11_banking_data.csv')
data
```

```python
#sorting data based on account balance (in descending order)
sorted_data = data.sort_values(by="Account_Balance",ascending=False)
sorted_data
```

```python
"""
You can use the rank() function in Pandas to rank transactions based on Transaction_Amount within each branch.
 The groupby() method is used to group the data by branch, and the rank() method is applied to assign ranks within each branch.
"""
```

```python
data['Rank']= data.groupby("Branch")["Transaction_Amount"].rank(ascending=False)
```
