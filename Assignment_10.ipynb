```python
data = pd.read_csv('Day_10_banking_data.csv')
data
```

```python
data_filtered=data[data['Transaction_Amount'] > 2000]
data_filtered
```

```python
#Find all rows where the Transaction_Type is "Loan Payment" and the Account_Balance is greater than 5000.
balance_greater_than_5000 = data[data['Account_Balance']>5000]
balance_greater_than_5000
```

```python
#extracting transactions in the uptown branch
uptown_branch_transactions = data[data['Branch']=='Uptown']
uptown_branch_transactions
```

```python
#Data Transformation
```

```python
#adding a new column "Transaction_Fee" which is 2% of transaction amount.
data['Transaction_Fee'] = data['Transaction_Amount'] * 0.02
data
```

```python
#creating a new column balance status if acc balance > 5000
data['Balance_Status']= data['Account_Balance'].apply(lambda x: 'High' if x > 5000 else 'Low')
data
```
