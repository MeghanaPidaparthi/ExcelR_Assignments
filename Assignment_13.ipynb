```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Day_13_Pharma_data.csv')
data
```

```python
#checking for missing values
data.isnull().sum()
```

```python
data.dropna(inplace=True) #another metod to handle missing values
data
```

```python
#A bar plot showing total sales per region.
sns.barplot(x='Region',y='Sales',data=data,)
plt.xlabel('Region')
plt.ylabel('Sales')
plt.title('Sales per Region')
plt.show()
```

```python
#A scatter plot to visualize the relationship between Marketing_Spend and Sales.
sns.scatterplot(x=data['Sales'],y=data['Marketing_Spend'],palette=(['violet','red','blue']))
plt.xlabel('Sales')
plt.ylabel('Marketing Spend')
plt.title('Sales vs Marketing Spend')
plt.show()
```

```python
#A boxplot comparing drug effectiveness across different age groups.
sns.boxplot(x=data['Effectiveness'],y=data['Age_Group'])
plt.xlabel('Effectiveness')
plt.ylabel('Age Group')
plt.title('Effectiveness by Age Group')
plt.show()
```

```python
#A line plot showing the sales trend for each product over different trial periods.
sns.lineplot(x=data['Sales'],y=data['Trial_Period'])
```

```python
#A heatmap of the correlation between Sales, Marketing_Spend, and Effectiveness.
corr = data[['Sales','Marketing_Spend','Effectiveness']].corr()
sns.heatmap(corr,annot=True)
plt.show()
```
